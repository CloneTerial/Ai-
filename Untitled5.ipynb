```python
from yt_dlp import YoutubeDL

def download_video_ytdlp(url, download_path='/content'):
    try:
        # Konfigurasi untuk mengunduh video dengan resolusi terbaik
        ydl_opts = {
            'outtmpl': f'{download_path}/%(title)s.%(ext)s',  # Simpan video di path tertentu
            'format': 'best',  # Pilih kualitas terbaik
        }
        with YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        print("Download completed!")
    except Exception as e:
        print(f"Error: {e}")

# Input URL
youtube_url = input("Enter the YouTube video URL: ")
download_video_ytdlp(youtube_url)
```

```python
import yt_dlp
import IPython.display as display
from IPython.display import Image

# Fungsi untuk menampilkan metadata dan thumbnail
def display_metadata(info_dict):
    print(f"Title: {info_dict['title']}")
    print(f"Uploader: {info_dict['uploader']}")
    print(f"Duration: {info_dict['duration']} seconds")
    print(f"View Count: {info_dict['view_count']}")
    print(f"Uploader URL: {info_dict['uploader_url']}")
    print(f"Thumbnail: {info_dict['thumbnail']}")
    display.display(Image(info_dict['thumbnail']))

    # Menampilkan resolusi dan format yang tersedia (tidak ada duplikat)
    print("\nAvailable formats:")
    seen_resolutions = set()  # Menghindari resolusi duplikat
    for f in info_dict['formats']:
        if f.get('height'):
            resolution = f['height']
            if resolution not in seen_resolutions:
                print(f"Format: | Resolution: {resolution}p | Extension: {f['ext']}")
                seen_resolutions.add(resolution)
        elif f['ext'] == 'mp3':  # Menampilkan format audio MP3
            print(f"Format:  | Audio | Extension: {f['ext']}")

# Fungsi untuk mendownload video atau audio berdasarkan pilihan format dan resolusi
def download_video_or_audio(url, format_choice, resolution_choice=None):
    ydl_opts = {
        'noplaylist': True,  # Hanya download satu video, bukan playlist
        'outtmpl': '%(title)s.%(ext)s',  # Nama file output
        'quiet': False
    }

    # Jika memilih format audio MP3
    if format_choice == 'mp3':
        ydl_opts['format'] = 'bestaudio/best'
        ydl_opts['postprocessors'] = [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }]
    # Jika memilih format video MP4 dengan resolusi tertentu
    elif format_choice == 'mp4' and resolution_choice:
        ydl_opts['format'] = f'bestvideo[height={resolution_choice}]+bestaudio/best'

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(url, download=False)
        display_metadata(info_dict)  # Tampilkan metadata dan thumbnail
        ydl.download([url])

# Input URL dari pengguna
url = input("Masukkan URL YouTube: ")

# Pilihan format: MP3 (audio) atau MP4 (video)
print("\nPilih format:")
print("1. MP3 (Audio terbaik)")
print("2. MP4 (Video terbaik dengan resolusi pilihan)")
format_choice_index = int(input("\nMasukkan pilihan format (1/2): "))
format_choice = 'mp3' if format_choice_index == 1 else 'mp4'

# Jika memilih MP4, tampilkan pilihan resolusi
if format_choice == 'mp4':
    with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
        info_dict = ydl.extract_info(url, download=False)
        available_resolutions = [f'{f["height"]}p' for f in info_dict['formats'] if f.get('height')]

    print("\nPilih resolusi video:")
    for i, resolution in enumerate(available_resolutions, 1):
        print(f"{i}. {resolution}")

    resolution_choice_index = int(input("\nMasukkan pilihan resolusi (1, 2, 3, ...): ")) - 1
    resolution_choice = int(available_resolutions[resolution_choice_index].replace('p', ''))  # Ambil angka resolusi
else:
    resolution_choice = None  # Tidak ada resolusi jika memilih MP3

# Download berdasarkan pilihan format dan resolusi
download_video_or_audio(url, format_choice, resolution_choice)
```

```python
import cv2
import mediapipe as mp
from google.colab import files
from IPython.display import display, HTML
import time
import os
from base64 import b64encode

# ðŸ”¹ Inisialisasi MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=2,
                       min_detection_confidence=0.5,
                       min_tracking_confidence=0.5)
mp_draw = mp.solutions.drawing_utils

# Warna oranye dan ungu (BGR)
color_orange = (0, 165, 255)  # Oranye dalam format BGR
color_purple = (128, 0, 128)  # Ungu dalam format BGR

# ðŸ”¹ Unggah video
uploaded = files.upload()
video_filename = list(uploaded.keys())[0]  # Ambil nama file pertama yang diunggah

# ðŸ”¹ Buka video
cap = cv2.VideoCapture(video_filename)

# ðŸ”¹ Periksa apakah video terbuka dengan benar
if not cap.isOpened():
    print("Error: Tidak dapat membuka video.")
    cap.release()
    exit()

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = cap.get(cv2.CAP_PROP_FPS) or 30  # Gunakan 30 FPS jika tidak terdeteksi

# ðŸ”¹ Gunakan codec yang lebih kompatibel (XVID)
output_filename = 'output.mp4'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_filename, fourcc, fps, (frame_width, frame_height))

frame_count = 0
start_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Hentikan jika tidak ada frame

    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Konversi warna agar benar
    results = hands.process(frame)

    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            # Gambar titik (landmark) dengan warna ungu
            for lm in handLms.landmark:
                cx, cy = int(lm.x * frame_width), int(lm.y * frame_height)
                cv2.circle(frame, (cx, cy), 6, color_purple, -1)  # Titik ungu

            # Gambar koneksi tangan dengan warna oranye
            mp_draw.draw_landmarks(
                frame, handLms, mp_hands.HAND_CONNECTIONS,
                mp_draw.DrawingSpec(color=color_orange, thickness=3, circle_radius=2),  # Garis oranye
                mp_draw.DrawingSpec(color=color_purple, thickness=2, circle_radius=4))  # Titik ungu

    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)  # Konversi kembali ke BGR untuk OpenCV
    out.write(frame)  # Simpan frame ke video output
    frame_count += 1

cap.release()
out.release()

# ðŸ”¹ Hitung waktu eksekusi
end_time = time.time()
print(f" {frame_count} {end_time - start_time:.2f} detik.")

# ðŸ”¹ Konversi agar lebih kompatibel
os.system("ffmpeg -i output.mp4 -vcodec libx264 fixed_output.mp4 -y")

# ðŸ”¹ Tampilkan hasil video di Colab
def show_video(video_path, width=600):
    with open(video_path, 'rb') as f:
        video_data = f.read()
        video_base64 = b64encode(video_data).decode()
    video_tag = f'''
    <video width="{width}" controls>
        <source src="data:video/mp4;base64,{video_base64}" type="video/mp4">
        Maaf, browser Anda tidak mendukung video HTML5.
    </video>
    '''
    return HTML(video_tag)

# Tampilkan video hasil
show_video('fixed_output.mp4', width=250)
```

```python
import cv2
import mediapipe as mp
from google.colab import files
from IPython.display import display, HTML
import time
import os
from base64 import b64encode

# Fungsi untuk menghitung jari yang terangkat pada satu tangan
def count_fingers(hand_landmarks, hand_label="Right"):
    """
    Menghitung jumlah jari yang terangkat.
    Finger tip indices: thumb=4, index=8, middle=12, ring=16, pinky=20.

    Untuk jari selain ibu jari, jika posisi landmark tip (misalnya, landmark 8 untuk jari telunjuk)
    lebih tinggi (y lebih kecil) daripada landmark PIP (misalnya, landmark 6 untuk jari telunjuk)
    maka jari tersebut dianggap terangkat.

    Untuk ibu jari, kita gunakan perbandingan pada sumbu-x,
    dengan asumsi bahwa untuk tangan kanan: jika landmark 4 lebih kiri (x lebih kecil) dari landmark 3, maka terangkat;
    sedangkan untuk tangan kiri, kondisinya dibalik.
    """
    finger_tips_ids = [4, 8, 12, 16, 20]
    fingers = []

    # Ibu jari
    if hand_label == "Right":
        if hand_landmarks.landmark[4].x < hand_landmarks.landmark[3].x:
            fingers.append(1)
        else:
            fingers.append(0)
    else:  # Left hand
        if hand_landmarks.landmark[4].x > hand_landmarks.landmark[3].x:
            fingers.append(1)
        else:
            fingers.append(0)

    # Empat jari lainnya (index, middle, ring, pinky)
    for tip_id in finger_tips_ids[1:]:
        if hand_landmarks.landmark[tip_id].y < hand_landmarks.landmark[tip_id - 2].y:
            fingers.append(1)
        else:
            fingers.append(0)

    return sum(fingers)

# Inisialisasi MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=2,
                       min_detection_confidence=0.5,
                       min_tracking_confidence=0.5)
mp_draw = mp.solutions.drawing_utils

# Warna yang digunakan (format BGR)
color_orange = (0, 165, 255)  # Oranye untuk koneksi garis
color_purple = (128, 0, 128)  # Ungu untuk titik landmark

# Unggah video dari komputer
uploaded = files.upload()
video_filename = list(uploaded.keys())[0]  # Ambil nama file pertama yang diupload

# Buka video
cap = cv2.VideoCapture(video_filename)

# Periksa apakah video terbuka dengan benar
if not cap.isOpened():
    print("Error: Tidak dapat membuka video.")
    cap.release()
    exit()

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS) or 30  # Gunakan 30 FPS jika tidak terdeteksi

# Inisialisasi VideoWriter dengan codec yang kompatibel (XVID)
output_filename = 'output.mp4'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_filename, fourcc, fps, (frame_width, frame_height))

frame_count = 0
start_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Hentikan jika tidak ada frame

    # Konversi frame dari BGR ke RGB untuk MediaPipe
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(frame_rgb)

    if results.multi_hand_landmarks:
        # Jika terdapat lebih dari satu tangan, proses masing-masing tangan
        for idx, handLms in enumerate(results.multi_hand_landmarks):
            # Dapatkan label tangan (Right/Left) jika tersedia
            hand_label = "Right"
            if results.multi_handedness:
                hand_label = results.multi_handedness[idx].classification[0].label

            # Hitung jumlah jari yang terangkat
            finger_count = count_fingers(handLms, hand_label)

            # Gambar titik landmark dengan warna ungu
            for lm in handLms.landmark:
                cx, cy = int(lm.x * frame_width), int(lm.y * frame_height)
                cv2.circle(frame, (cx, cy), 6, color_purple, -1)

            # Gambar koneksi antar landmark dengan warna oranye
            mp_draw.draw_landmarks(frame, handLms, mp_hands.HAND_CONNECTIONS,
                                   mp_draw.DrawingSpec(color=color_orange, thickness=3, circle_radius=2),
                                   mp_draw.DrawingSpec(color=color_purple, thickness=5, circle_radius=4))

            # Tempatkan teks jumlah jari di dekat pergelangan tangan (landmark 0)
            x0 = int(handLms.landmark[0].x * frame_width)
            y0 = int(handLms.landmark[0].y * frame_height)
            cv2.putText(frame, str(finger_count), (x0, y0 - 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 255), 3)

    # Tulis frame yang telah diproses ke video output
    out.write(frame)
    frame_count += 1

cap.release()
out.release()

end_time = time.time()
print(f" {frame_count} frame {end_time - start_time:.2f} detik.")

# Konversi video output agar lebih kompatibel (menggunakan ffmpeg)
os.system("ffmpeg -i output.mp4 -vcodec libx264 fixed_output.mp4 -y")

# Fungsi untuk menampilkan video di Colab
def show_video(video_path, width=600):
    with open(video_path, 'rb') as f:
        video_data = f.read()
        video_base64 = b64encode(video_data).decode()
    video_tag = f'''
    <video width="{width}" controls>
        <source src="data:video/mp4;base64,{video_base64}" type="video/mp4">
        Maaf, browser Anda tidak mendukung video HTML5.
    </video>
    '''
    return HTML(video_tag)

# Tampilkan video hasil
show_video('fixed_output.mp4', width=220)
```

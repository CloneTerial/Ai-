```python
import cv2
import mediapipe as mp
from google.colab import files
import os

# Inisialisasi MediaPipe
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Upload video ke Colab
uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# Buka video
cap = cv2.VideoCapture(video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Simpan output video
output_path = 'output.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(frame_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    out.write(frame)

cap.release()
out.release()

# Tampilkan hasil video
from IPython.display import display, HTML
display(HTML(f'<video controls src="{output_path}" width="640"></video>'))
```

```python
from IPython.display import HTML
from base64 import b64encode

# Ubah video menjadi base64 untuk ditampilkan
def show_video(video_path):
    with open(video_path, "rb") as video_file:
        video_base64 = b64encode(video_file.read()).decode()
    return HTML(f'<video controls src="data:video/mp4;base64,{video_base64}" width="640"></video>')

# Tampilkan video
show_video("output.mp4")  # Ganti dengan nama file yang sesuai
```

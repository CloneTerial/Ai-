```python
import requests
from bs4 import BeautifulSoup
import time

# Variabel global untuk menyimpan ID berita terakhir
last_news_id = None

# Fungsi untuk scrape berita terbaru
def fetch_latest_news():
    global last_news_id

    # URL CNBC Indonesia
    url = "https://www.cnbcindonesia.com/news"
    headers = {'User-Agent': 'Mozilla/5.0'}

    try:
        # Fetch halaman utama
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Cek apakah request berhasil

        # Parsing HTML
        soup = BeautifulSoup(response.content, 'html.parser')

        # Temukan elemen div dengan class `nhl-list`
        nhl_list = soup.find('div', class_='nhl-list')
        if nhl_list:
            # Cari semua elemen <article> di dalam `nhl-list`
            articles = nhl_list.find_all('article')

            # Iterasi melalui semua artikel
            for article in articles:
                # Ambil link berita
                link = article.find('a')['href'] if article.find('a') else None

                # Ambil judul berita
                title = article.find('h2').text.strip() if article.find('h2') else "Tidak ada judul"

                # Ambil kategori
                category = article.find('span', class_='mb-1').text.strip() if article.find('span', class_='mb-1') else "Tidak ada kategori"

                # Ambil waktu publikasi
                time_posted = article.find('span', class_='text-xs').text.strip() if article.find('span', class_='text-xs') else "Tidak ada waktu"

                # Ambil URL gambar
                image_url = article.find('img')['src'] if article.find('img') else "Tidak ada gambar"

                # Ambil ID berita (jika ada atribut dtr-id)
                news_id = article.find('a').get('dtr-id', None)

                # Cek apakah berita ini sudah dikirim sebelumnya
                if news_id and news_id != last_news_id:
                    last_news_id = news_id  # Update ID berita terakhir
                    print(f"Berita Terbaru:")
                    print(f"Judul       : {title}")
                    print(f"Kategori    : {category}")
                    print(f"Waktu       : {time_posted}")
                    print(f"Link        : {link}")
                    print(f"Gambar      : {image_url}\n")
                else:
                    print(f"Tidak ada berita baru untuk ID {news_id}\n")
        else:
            print("Div dengan class `nhl-list` tidak ditemukan!\n")

    except Exception as e:
        print(f"Terjadi kesalahan: {e}\n")

# Fungsi untuk menjalankan loop scraping (tanpa bot)
def run_scraping_test():
    print("Memulai pengujian scraping...\n")
    try:
        for i in range(5):  # Lakukan pengujian sebanyak 5 iterasi (dengan delay)
            print(f"Iterasi ke-{i + 1}:")
            fetch_latest_news()
            time.sleep(300)  # Tunggu 1 menit sebelum iterasi berikutnya
    except KeyboardInterrupt:
        print("\nPengujian dihentikan oleh pengguna.")
    finally:
        print("Pengujian selesai.")

# Jalankan pengujian scraping
run_scraping_test()
```

```python
import requests
from bs4 import BeautifulSoup
import schedule
import time

# Variabel global untuk menyimpan ID berita terakhir
last_news_id = None

# Fungsi untuk scrape berita terbaru
def fetch_latest_news():
    global last_news_id

    # URL CNBC Indonesia
    url = "https://www.cnbcindonesia.com/"
    headers = {'User-Agent': 'Mozilla/5.0'}

    # Fetch halaman utama
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Temukan elemen artikel pertama (berita terbaru)
        article = soup.find('article')
        if article:
            # Ambil ID berita dari atribut `dtr-id`
            news_id = article.find('a').get('dtr-id', None)
            title = article.find('h2').text.strip()
            link = article.find('a')['href']

            # Kirim berita jika ID berbeda dari ID terakhir
            if news_id != last_news_id:
                last_news_id = news_id
                print(f"Berita Terbaru: {title}")
                print(f"Link: {link}\n")
                # Tambahkan logika untuk mengirim berita ke bot di sini (misalnya Telegram API)
        else:
            print("Tidak ada berita ditemukan!")
    else:
        print("Gagal mengakses website CNBC Indonesia")

# Jadwalkan scraping setiap menit
schedule.every(5).minutes.do(fetch_latest_news)

print("Bot siap mengirim berita terbaru setiap menit...")

# Looping untuk menjalankan schedule
while True:
    schedule.run_pending()
    time.sleep(1)
```
